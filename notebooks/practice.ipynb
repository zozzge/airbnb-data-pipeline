```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("/Users/zeynep/Desktop/python-sql-practice/database/practice.d")
cursor = conn.cursor()

cursor.execute(""" 
CREATE TABLE IF NOT EXISTS employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    salary INTEGER,
    department TEXT
)
""")

cursor.execute("INSERT INTO employees (name, salary, department) VALUES ('Ali', 50000, 'IT')")
cursor.execute("INSERT INTO employees (name, salary, department) VALUES ('Ayşe', 65000, 'HR')")
cursor.execute("INSERT INTO employees (name, salary, department) VALUES ('Mehmet', 72000, 'Finance')")

conn.commit()

query = "SELECT * FROM employees"
df = pd.read_sql(query, conn)
df
```

```python
import sqlite3
import pandas as pd
import os

db_path = "/Users/zeynep/Desktop/python-sql-practice/database/practice.db"
print("DB yolu:", os.path.abspath(db_path))

conn = sqlite3.connect(db_path)
cursor = conn.cursor()

tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("Tablolar:")
print(tables)

df = pd.read_sql("SELECT id, name, salary, department FROM employees ORDER BY id;", conn)
print("Employees tablosu:")
print(df)

conn.close()
```

```python
import sqlite3
import pandas as pd

db_path = "/Users/zeynep/Desktop/python-sql-practice/database/practice.db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

cursor.execute("DROP TABLE IF EXISTS employees")

cursor.execute("""
CREATE TABLE employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    salary INTEGER,
    department TEXT
)
""")

cursor.execute("INSERT INTO employees (name, salary, department) VALUES ('Ali', 5000, 'IT')")
cursor.execute("INSERT INTO employees (name, salary, department) VALUES ('Ayşe', 6500, 'HR')")
cursor.execute("INSERT INTO employees (name, salary, department) VALUES ('Mehmet', 7200, 'Finance')")

conn.commit()

df = pd.read_sql("SELECT * FROM employees", conn)
print(df)

conn.close()
```
