```python
df["rating_level"] = np.where(df["Aggregate rating"] >= 4.5, "High", "Low")

# Rating seviyesine göre yeni sütun
```

```python
df["popularity"] = np.select(
    [
        (df["Votes"] >= 1000) & (df["Aggregate rating"] >= 4.5),
        (df["Votes"] >= 500)
    ],
    [
        "Very popular",
        "Popular"
    ],
    default="Normal"

)

df[["Restaurant Name", "Aggregate rating", "Votes", "rating_level", "popularity"]].head()
```

```python
pd.crosstab(df["City"], df["rating_level"])

# Şehir ve rating dağılımı
```

```python
pd.pivot_table(
    df,
    values="Aggregate rating",
    index="City",
    aggfunc="mean"
).sort_values("Aggregate rating", ascending=False)

# Ortalama rating pivot
```

```python
df[["Aggregate rating", "Votes", "Average Cost for two"]].corr()

# Sayısal sütunların ilişkisini gösterir
```

```python
df["Aggregate rating"].hist(bins=20)
plt.title("Rating Distribution")
plt.xlabel("Aggregate rating")
plt.ylabel("Frequency")
plt.show()

# Histogram
```

```python
df["City"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Cities by Restaurant Count")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()

# Şehir bazında restoran sayısı
```

```python
# Ortalama rating bar chart

df.groupby("City")["Aggregate rating"].mean().sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Top 10 Cities by Average Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.show()
```
